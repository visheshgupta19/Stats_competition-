```python
#Processing dates
def processDates(df, start_year=2024):

    df['Date1'] = pd.to_datetime(df['Date'].str[4:], format='%m-%d')
    df.drop(['Date'], inplace=True, axis=1)
    
    # Add the year dynamically, incrementing after December
    year = start_year
    df['Date'] = df['Date1'].apply(lambda date: date.replace(year=year if date.month >= (df['Date1'].iloc[0].month if 'Date1' in df else 1) else year + 1))
    
    return df

#write specific columns to df
def writeToCSV(df):
    
    columns_to_keep = ['Date', 'Team', 'Opponent', 'Location', 
                       'WAB', 'ADJO', 'ADJD', 'EFF', 'EFG%', 'TO%', 
                       'OR%', 'FTR', '2P', '3P', 'Opp EFF', 'Opp EFG%', 
                       'Opp TO%', 'Opp OR%', 'Opp FTR', 'Opp 2P', 'Opp 3P', 
                       'Opponent_score', 'Team_score']
    
    df = df[columns_to_keep]
    df.to_csv('/files/team_data_collected.csv', index=False)
    
    return "Done"

dfProcessed = processDates(combined_df)
dfForR = writeToCSV(dfProcessed)
```
