```python
import pandas as pd
```

```python
df_main = pd.read_csv("/files/team_data_collected.csv")
df_opponent = pd.read_csv("/files/only_opponent_data.csv")
```

```python
df_opponent = df_opponent.rename(columns={"date": "Date", 
                                          "team": "Opponent", "opp": "Team",
                                          "adj_o": "opp_adj_o", "adj_d": "opp_adj_d"})
df_opponent.head(20)
```

```python
# Merge the two dataframes on 'date', 'team', and 'opp'
merged_df = pd.merge(df_main, df_opponent, on=['Date', 'Team', 'Opponent'])

# Display the merged DataFrame
display(merged_df)

merged_df.to_csv('final_opponent_and_team_data.csv', index=False)
```
